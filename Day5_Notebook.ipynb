## Task 1: Line & Scatter Plots

### Task 1.1 - Line Plot with Custom Style

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.arange(1, 11)
y = x ** 2

plt.plot(x, y,
         linestyle='--',
         color='green',
         linewidth=2)

plt.show()
```

### Task 1.2 - Multiple Lines with Legend

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 11)

y1 = x ** 2
y2 = x ** 3

plt.plot(x, y1, label='y = x²')
plt.plot(x, y2, label='y = x³')

plt.legend()

plt.show()
```

### Task 1.3 - Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np

height = np.array([150, 155, 160, 165, 170, 175, 180, 185])
weight = np.array([50, 55, 60, 65, 70, 75, 80, 85])

plt.scatter(
    height,
    weight,
    color='purple',
    marker='^'
)

plt.xlabel('Height')
plt.ylabel('Weight')

plt.show()
```

## Task 2: Bar, Histogram, Pie

### Task 2.1 - Horizontal Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np

countries = ['India', 'USA', 'China', 'Brazil', 'Canada']

population = [1400, 331, 1440, 213, 38]

plt.barh(countries, population)

plt.xlabel('Population (Millions)')
plt.ylabel('Countries')

plt.show()
```

### Task 2.2 - Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

ages = np.random.randint(0, 81, 500)

plt.hist(
    ages,
    bins=10,
    edgecolor='black'
)

plt.xlabel('Age')
plt.ylabel('Frequency')

plt.show()
```

### Task 2.3 - Pie Chart

```python
import matplotlib.pyplot as plt
import numpy as np

activities = ['Sleep', 'Work', 'Study', 'Leisure']

hours = [8, 8, 4, 4]

explode = [0, 0, 0.2, 0]

plt.pie(
    hours,
    labels=activities,
    autopct='%1.1f%%',
    explode=explode
)

plt.show()
```

## Task 3: Customization & Subplots

### Task 3.1 - Customized Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 11)

y = x ** 2

plt.plot(
    x,
    y,
    linestyle='--',
    color='green'
)

plt.title('Squares of Numbers')

plt.xlabel('Numbers')

plt.ylabel('Squares')

plt.xlim(0, 12)

plt.grid(True)

plt.show()
```

### Task 3.2 - Subplots Grid

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

x = np.arange(1, 11)

axes[0, 0].plot(x, x ** 2)
axes[0, 0].set_title('Line Plot')

axes[0, 1].scatter(x, np.random.randint(1, 100, 10))
axes[0, 1].set_title('Scatter Plot')

categories = ['A', 'B', 'C']

values = [10, 20, 15]

axes[1, 0].bar(categories, values)
axes[1, 0].set_title('Bar Chart')

data = np.random.randn(1000)

axes[1, 1].hist(data, bins=20)
axes[1, 1].set_title('Histogram')

plt.tight_layout()

plt.show()
```

### Task 3.3 - Save Subplot to File

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

x = np.arange(1, 11)

axes[0, 0].plot(x, x ** 2)
axes[0, 0].set_title('Line Plot')

axes[0, 1].scatter(x, np.random.randint(1, 100, 10))
axes[0, 1].set_title('Scatter Plot')

categories = ['A', 'B', 'C']

values = [10, 20, 15]

axes[1, 0].bar(categories, values)
axes[1, 0].set_title('Bar Chart')

data = np.random.randn(1000)

axes[1, 1].hist(data, bins=20)
axes[1, 1].set_title('Histogram')

plt.tight_layout()

plt.savefig(
    'subplot_grid.jpg',
    dpi=300
)

plt.show()
```
