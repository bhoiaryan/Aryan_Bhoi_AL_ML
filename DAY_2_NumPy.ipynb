## Section 1 — Advanced Indexing & Filtering

### Task 1.4.1 — Filter divisible by 3 and 5

```python
import numpy as np

a = np.random.randint(1, 100, (6, 6))

d = a[(a % 3 == 0) & (a % 5 == 0)]

print(d)
```

### Task 1.4.2 — Boolean mask: above mean → 1, else 0

```python
import numpy as np

a = np.random.randint(1, 100, (3, 3))

l = np.where(a > np.mean(a), 1, 0)
print(l)
```

### Task 1.4.3 — Fancy indexing with np.ix_

```python
import numpy as np

a = np.arange(25).reshape((5, 5))

rows = [1, 3, 4]
cols = [0, 2, 4]

sub = a[np.ix_(rows, cols)]
print(sub)
```

### Task 1.4.4 — Sort rows and columns independently

### Task 1.4.5 — Grade assignment with np.select

```python
import numpy as np

a = np.random.randint(1, 100, 15)

scores = [a >= 85, a >= 70, a >= 55, a >= 40]
grades = ['A', 'B', 'C', 'D']

print(np.select(scores, grades, default='F'))
```

### Task 2.2.2 — Column-wise normalization

```python
import numpy as np

a = np.random.randint(1, 100, (6, 4))

c_mean = a.mean(axis=0)
c_std = a.std(axis=0)

normalised = (a - c_mean) / (c_std)

print(normalised)
print(normalised.mean())
print(normalised.std())
```

### Task 3.3.1 — Matrix inverse verification

```python
import numpy as np

a = np.random.randint(1, 100, (3, 3)).astype(float)

ai = np.linalg.inv(a)
p = a @ ai

print(np.allclose(p, np.eye(3)))
```

```python
import numpy as np

a = np.random.randint(1, 100, (3, 3))

s = a @ a.T
print(s)

vals, vecs = np.linalg.eig(s)
print(vals)
print(vecs)
```

## Section 6 — Performance & Advanced

### Task 6.3.1 — Loop vs vectorization timing

```python
import numpy as np
import time

x = np.array([1, 2, 3, 4, 5])

start = time.time()
result = []
for i in range(len(x)):
    result.append(x[i] ** 2 + 2 * x[i] + 1)
end = time.time()
print("Loop: ", end - start)

t = time.perf_counter()
result = x ** 2 + 2 * x + 1
print("Vectorisation: ", time.perf_counter() - t)
```

### Task 6.3.2 — einsum outer product

```python
import numpy as np

a = np.array([1, 2, 3, 4, 5])
b = np.array([6, 7, 8, 9, 5])

C = np.einsum('i,j->ij', a, b)

print(np.allclose(C, np.outer(a, b)))
```

### Task 6.3.3 — float32 vs float64 memory & speed

```python
import numpy as np
import time

A = np.random.rand(1_000_000).astype(np.float32)
B = np.random.rand(1_000_000).astype(np.float64)

print("A (float32) bytes:", A.nbytes)
print("B (float64) bytes:", B.nbytes)

start = time.time()
sq32 = A ** 2
t32 = time.time() - start

start = time.time()
sq64 = B ** 2
t64 = time.time() - start

print("float32 time:", t32)
print("float64 time:", t64)
```

### Challenge 7.1 — Stats report function

```python
import numpy as np

def stats_report(arr):
    print('Shape: ', np.shape(arr))
    print('Min: ', np.min(arr))
    print('Max: ', np.max(arr))
    print('Mean: ', np.mean(arr))
    print('Median: ', np.median(arr))
    print('STD: ', np.std(arr))
    print('25th and 75th percentile', np.percentile(arr, 25), np.percentile(arr, 75))
    r = np.unravel_index(np.argmax(arr), arr.shape)
    print("Max position: ", r)

stats_report(np.random.rand(20, 5))
```

### Challenge 7.3 — Student ranking & correlation

```python
import numpy as np

marks = np.random.randint(30, 101, (30, 5))

avg = marks.mean(axis=1)
ranks = avg.argsort()[::-1].argsort() + 1
top5 = np.argsort(avg)[-5:]
bottom5 = np.argsort(avg)[:5]
corr = np.corrcoef(marks[:, 0], marks[:, 1])

print("Ranks:", ranks)
print("Top 5:", top5)
print("Bottom 5:", bottom5)
print("Correlation matrix:\n", corr)
```

### Challenge 7.4 — Monte Carlo Pi estimation

```python
import numpy as np

N = 1_000_000

points = np.random.rand(N, 2)
inside = np.sum(points[:, 0] ** 2 + points[:, 1] ** 2 <= 1)
pi_estimate = 4 * inside / N

print("Estimated Pi:", pi_estimate)
```

### Challenge 7.5 — Portfolio risk analysis

```python
import numpy as np

returns = np.random.rand(12, 5)

mean_returns = returns.mean(axis=0)
cov_matrix = np.cov(returns, rowvar=False)
weights = np.array([0.2] * 5)
portfolio_variance = weights @ cov_matrix @ weights
asset_risk = returns.std(axis=0)
highest_risk = np.argmax(asset_risk)
lowest_risk = np.argmin(asset_risk)

print("Mean returns:", mean_returns)
print("Covariance matrix:\n", cov_matrix)
print("Portfolio variance:", portfolio_variance)
print("Highest risk asset:", highest_risk)
print("Lowest risk asset:", lowest_risk)
```
