```python
import numpy as np

a = np.random.randint(1, 10, 10)

mean = a.mean()
std = a.std()

norm = (a - mean) / std
print(norm)

print("STD :", norm.std())
print('Mean: ', norm.mean())
```

## Section 7 — Comprehensive Challenges

### Challenge 7.1 — Student marks analysis

```python
import numpy as np

marks = np.array([[72, 85, 60, 78], [88, 92, 95, 90], [45, 52, 48, 55], [65, 70, 72, 68]])

# a
print("Total :", marks.sum(axis=1))
print("Average:", marks.mean(axis=1))

# b
print("Subject Average:", marks.mean(axis=0))

# c
avg = marks.mean(axis=1)
print(np.where(avg >= 80, "Dist", np.where(avg >= 65, "Merit", np.where(avg >= 50, 'Pass', "Fail"))))
```

### Challenge 7.2 — Sales analysis

```python
import numpy as np

sales = np.random.randint(100, 1000, (12, 4))

monthly_total = sales.sum(axis=1)
best_month = monthly_total.argmax()
product_total = sales.sum(axis=0)
worst_product = product_total.argmin()
months = np.where((sales > 500).all(axis=1))[0]

print("Best Month:", best_month)
print("Worst Product:", worst_product)
print("Months >500:", months)
```

### Challenge 7.3 — Temperature normalization

```python
import numpy as np

temp = np.random.randint(15, 41, (7, 24))

norm = (temp - temp.min(axis=1, keepdims=True)) / (
       temp.max(axis=1, keepdims=True) -
       temp.min(axis=1, keepdims=True)
)
print(norm.min(axis=1))
print(norm.max(axis=1))
```

### Challenge 7.4 — Custom my_range function

```python
import numpy as np

def my_range(start, stop, step):
    n = int((stop - start) / step)
    arr = np.zeros(n)
    value = start
    for i in range(n):
        arr[i] = value
        value += step
    return arr

print(my_range(0, 10, 2))
```

### Challenge 7.5 — 10x10 random matrix full analysis

```python
import numpy as np

matrix = np.random.rand(10, 10)

print("Shape:", matrix.shape)
print("Datatype:", matrix.dtype)
print("Total Sum:", matrix.sum())
print("Row Means:", matrix.mean(axis=1))
print("Column Std:", matrix.std(axis=0))

max_pos = np.unravel_index(matrix.argmax(), matrix.shape)
print("Global Max:", matrix.max())
print("Position:", max_pos)
print("Count > 0.5:", np.sum(matrix > 0.5))
```
