# Day 6 - Seaborn

This notebook contains all tasks from Day 6 covering scatter plots, line plots, histograms, boxplots, heatmaps, and pairplots.

## Task 1: Scatter & Line Plots

### Task 1.1 - Scatter Plot with Penguins

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

penguins = sns.load_dataset('penguins')

sns.scatterplot(
    data=penguins,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='species'
)

plt.title('Bill Length vs Bill Depth')

plt.show()
```

### Task 1.2 - Line Plot with Flights

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

flights = sns.load_dataset('flights')

sns.lineplot(
    data=flights,
    x='year',
    y='passengers'
)

plt.title('Passengers Over Years')

plt.show()
```

## Task 2: Distribution Plots

### Task 2.1 - Histogram with Titanic

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

titanic = sns.load_dataset('titanic')

sns.histplot(
    data=titanic,
    x='age',
    kde=True,
    hue='survived'
)

plt.title('Passenger Age Distribution')

plt.show()
```

### Task 2.2 - Count Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

titanic = sns.load_dataset('titanic')

sns.countplot(
    data=titanic,
    x='pclass'
)

plt.title('Passengers by Class')

plt.show()
```

### Task 2.3 - Box Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

titanic = sns.load_dataset('titanic')

sns.boxplot(
    data=titanic,
    x='pclass',
    y='fare'
)

plt.title('Fare Distribution by Passenger Class')

plt.show()
```

## Task 3: Heatmaps & Pairplots

### Task 3.1 - Correlation Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

tips = sns.load_dataset('tips')

numeric_df = tips.select_dtypes(include='number')

corr = numeric_df.corr()

plt.figure(figsize=(8, 6))

sns.heatmap(
    corr,
    annot=True,
    cmap='viridis'
)

plt.title('Tips Correlation Heatmap')

plt.show()
```

### Task 3.2 - Pairplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

iris = sns.load_dataset('iris')

sns.pairplot(
    iris,
    hue='species'
)

plt.show()
```

## Task 4: Diamonds Dataset Analysis

### Task 4a - Count Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

diamonds = sns.load_dataset('diamonds')

sns.countplot(
    data=diamonds,
    x='cut'
)

plt.title('Diamond Cut Types')

plt.show()
```

### Task 4b - Box Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

diamonds = sns.load_dataset('diamonds')

sns.boxplot(
    data=diamonds,
    x='color',
    y='price'
)

plt.title('Price Distribution by Color')

plt.show()
```

### Task 4c - Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

diamonds = sns.load_dataset('diamonds')

sns.scatterplot(
    data=diamonds,
    x='carat',
    y='price',
    hue='clarity'
)

plt.title('Carat vs Price')

plt.show()
```

### Task 4d - Correlation Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

diamonds = sns.load_dataset('diamonds')

numeric_df = diamonds.select_dtypes(include='number')

corr = numeric_df.corr()

plt.figure(figsize=(10, 8))

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm'
)

plt.title('Diamonds Correlation Heatmap')

plt.show()
```

### Task 4e - Save Plot to File

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

diamonds = sns.load_dataset('diamonds')

sns.scatterplot(
    data=diamonds,
    x='carat',
    y='price',
    hue='clarity'
)

plt.title('Diamonds Analysis')

plt.savefig(
    'diamonds_analysis.png',
    dpi=300
)

plt.show()
```
