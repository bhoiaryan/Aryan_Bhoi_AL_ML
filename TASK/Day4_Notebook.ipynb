## Section 1: Sorting & Ranking

### Task 1.1 - Sort by Fare

```python
import seaborn as sns

df = sns.load_dataset("titanic")

fare = df.sort_values('fare', ascending=False).head()

print("Descending: ")
print(fare)

highest = fare[['who','fare', 'class']].head(10)

print("Highest Payer:")
print(highest)
```

### Task 1.2 - Sort by Multiple Columns

```python
import seaborn as sns

df = sns.load_dataset("titanic")

asc = df.sort_values(['pclass','age'],ascending=[True,False])

print(asc[['pclass', 'age', 'who']].head(10))

youngest = df.loc[df.groupby('pclass')['age'].idxmin()]

print("Youngest in each class: ")
print(youngest[['pclass','age','who']])
```

### Task 1.3 - Ranking Methods

```python
import pandas as pd

students = pd.DataFrame({
    'name': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],
    'marks': [90, 85, 90, 70, 60, 85, 95, 95, 75, 70]
})

methods = ['average', 'min', 'max', 'first', 'dense']

for m in methods:
    students[f'rank_{m}'] = students['marks'].rank(ascending=False, method=m)

print(students)
```

## Section 2: GroupBy Operations

### Task 2.1 - GroupBy Survival Rate

```python
import seaborn as sns

df = sns.load_dataset("titanic")

print(df.groupby('sex')['survived'].mean())

print(df.groupby('pclass')['survived'].mean())


combo = df.groupby(['sex', 'pclass'])['survived'].mean()

print(combo)

print("Highest Survival Rate:")
print(combo.idxmax(), combo.max())
```

### Task 2.2 - GroupBy Aggregations

```python
import seaborn as sns

df = sns.load_dataset("titanic")

fare_stats = df.groupby('pclass')['fare'].agg(
    ['mean', 'median', 'std']
)

print(fare_stats)
```

### Task 2.3 - Transform

```python
import seaborn as sns

df = sns.load_dataset("titanic")
df['class_median_age'] = df.groupby('pclass')['age'].transform(
    'median'
)

df['IsOlderThanClassMedian'] = (
    df['age'] > df['class_median_age']
)

print(df[['age', 'pclass',
          'class_median_age',
          'IsOlderThanClassMedian']].head())
```

### Task 2.4 - Filter Groups

```python
import seaborn as sns

df = sns.load_dataset("titanic")
filtered = df.groupby('pclass').filter(
    lambda x: len(x) >= 200
)

print(filtered['pclass'].value_counts())

print("Remaining passengers:", len(filtered))
```

### Task 2.5 - Named Aggregations

```python
import seaborn as sns

df = sns.load_dataset("titanic")

result = df.groupby('embarked').agg(
    total_passengers=('survived', 'count'),
    survival_rate=('survived', 'mean'),
    avg_fare=('fare', 'mean')
).sort_values('survival_rate', ascending=False)

print(result)
```

## Section 3: Merging & Joining

### Task 3.1 - Merge Types

```python
import pandas as pd

students = pd.DataFrame({
    'id': [1,2,3,4],
    'name': ['Alice','Bob','Carol','Dave'],
    'age': [20,21,22,23]
})

grades = pd.DataFrame({
    'id': [1,2,3,5],
    'subject': ['Math','Sci','Eng','Math'],
    'marks': [90,80,85,70]
})


inner = pd.merge(students, grades, on='id', how='inner')


left = pd.merge(students, grades, on='id', how='left')

right = pd.merge(students, grades, on='id', how='right')

outer = pd.merge(students, grades, on='id', how='outer')

print(inner)
print(left)
print(right)
print(outer)
```

### Task 3.2 - Concat

```python
import pandas as pd

jan = pd.DataFrame({
    'Product': ['A','B'],
    'Sales': [100,200]
})

jan['Month'] = 'January'

feb = pd.DataFrame({
    'Product': ['A','B'],
    'Sales': [150,250]
})

feb['Month'] = 'February'

yearly = pd.concat([jan, feb], ignore_index=True)

print(yearly)
```

## Section 4: Pivot Tables & Crosstab

### Task 4.1 - Pivot Table

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset("titanic")

pivot = pd.pivot_table(
    df,
    values='age',
    index='pclass',
    columns='survived',
    aggfunc=['mean', 'std'],
    margins=True
)

print(pivot)
```

### Task 4.2 - Crosstab

```python
import seaborn as sns, pandas as pd


df = sns.load_dataset("titanic")

ct = pd.crosstab(
    [df['sex'], df['embarked']],
    df['survived'],
    normalize='index'
) * 100

print(ct)

print("Highest Survival:")
print(ct[1].idxmax())
```

### Task 4.3 - Pivot Revenue

```python
import seaborn as sns, pandas as pd


df = sns.load_dataset("titanic")

pivot = pd.pivot_table(
    df,
    values='fare',
    index='pclass',
    columns='embarked',
    aggfunc='sum'
)

print(pivot)

print("Highest revenue port per class:")
print(pivot.idxmax(axis=1))
```

### Task 5.1 - Map

### Task 5.2 - Apply

```python
import seaborn as sns, pandas as pd

df = sns.load_dataset("titanic")


def outcome(row):
    if row['survived'] == 1 and row['fare'] > 50:
        return 'Survived-Rich'
    elif row['survived'] == 1:
        return 'Survived-Poor'
    elif row['fare'] > 50:
        return 'Died-Rich'
    else:
        return 'Died-Poor'

df['Outcome'] = df.apply(outcome, axis=1)

print(df['Outcome'].value_counts())
```

### Task 5.3 - Cut & Qcut

```python
import seaborn as sns, pandas as pd

df = sns.load_dataset("titanic")

df['age_band'] = pd.cut(df['age'], bins=5)



df['age_quantile'] = pd.qcut(df['age'], q=5)

print(df['age_band'].value_counts())

print(df['age_quantile'].value_counts())
```

## Section 8: Mini Projects

### Task 8.1 - Sales Analysis

```python
import pandas as pd
import numpy as np

np.random.seed(1)

sales = pd.DataFrame({
    'Date': pd.date_range('2025-01-01', periods=50),
    'Salesperson': np.random.choice(
        ['Amit','Riya','Karan','Sneha'], 50),
    'Region': np.random.choice(
        ['North','South','East','West'], 50),
    'Product': np.random.choice(
        ['Laptop','Phone','Tablet'], 50),
    'Units': np.random.randint(1,20,50),
    'UnitPrice': np.random.randint(100,1000,50)
})

sales['Revenue'] = (
    sales['Units'] * sales['UnitPrice']
)

# GroupBy
summary = sales.groupby('Region').agg(
    total_revenue=('Revenue','sum'),
    total_units=('Units','sum'),
    avg_deal_size=('Revenue','mean')
)

# Pivot
pivot = pd.pivot_table(
    sales,
    values='Revenue',
    index='Region',
    columns='Product',
    aggfunc='sum'
)

# Ranking
rank = sales.groupby('Salesperson')['Revenue'].sum() \
            .rank(ascending=False)

print(summary)
print(pivot)
print(rank)

# Export
with pd.ExcelWriter('sales_analysis.xlsx') as writer:
    sales.to_excel(writer,
                   sheet_name='raw',
                   index=False)

    summary.to_excel(writer,
                     sheet_name='summary')

    pivot.to_excel(writer,
                   sheet_name='pivot')
```

### Task 8.2 - Titanic Full Analysis

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')


df['age'] = df['age'].fillna(df['age'].median())
df['embarked'] = df['embarked'].fillna(
    df['embarked'].mode()[0]
)

df['FamilySize'] = df['sibsp'] + df['parch'] + 1

df['IsAlone'] = (
    df['FamilySize'] == 1
).astype(int)

df['AgeGroup'] = pd.cut(
    df['age'],
    bins=[0,18,60,100],
    labels=['Child','Adult','Senior']
)

df['FareBand'] = pd.qcut(
    df['fare'],
    4,
    labels=['Low','Medium','High','Luxury']
)

ct = pd.crosstab(
    df['AgeGroup'],
    df['sex'],
    values=df['survived'],
    aggfunc='mean'
)

group = df.groupby('pclass').agg(
    mean_age=('age','mean'),
    mean_fare=('fare','mean'),
    survival_rate=('survived','mean'),
    passengers=('survived','count')
)

class_info = pd.DataFrame({
    'pclass':[1,2,3],
    'label':['First','Second','Third'],
    'deck_level':['Upper','Middle','Lower']
})

final = pd.merge(
    df,
    class_info,
    on='pclass',
    how='left'
)

final.to_csv('titanic_final.csv', index=False)

with pd.ExcelWriter('titanic_pivot.xlsx') as writer:
    ct.to_excel(writer,
                sheet_name='Crosstab')

    group.to_excel(writer,
                   sheet_name='GroupBy')
```

### Task 8.4 - Monte Carlo Simulation

```python
import numpy as np
import pandas as pd

n = 10000

df = pd.DataFrame({
    'round': range(1, n+1),
    'die1': np.random.randint(1,7,n),
    'die2': np.random.randint(1,7,n)
})

df['total'] = df['die1'] + df['die2']

freq = df.groupby('total').size()

prob = freq / n

print(prob)

p_gt9 = (df['total'] > 9).mean()

print("Probability > 9:", p_gt9)

pivot = pd.pivot_table(
    df,
    values='round',
    index='die1',
    columns='die2',
    aggfunc='count',
    fill_value=0
)

with pd.ExcelWriter('monte_carlo.xlsx') as writer:

    df.to_excel(writer,
                sheet_name='Raw',
                index=False)

    prob.to_excel(writer,
                  sheet_name='Stats')

    pivot.to_excel(writer,
                   sheet_name='Pivot')
```
