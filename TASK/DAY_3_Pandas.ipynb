## Section 2 — Pandas Series

### Task 2.2.1 — Temperature Series: hottest & coldest day

```python
import pandas as pd

temp = pd.Series([40, 42, 44, 41, 43, 43, 42],
                 index=['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun'])

hottest = temp.idxmax()
coldest = temp.idxmin()

print("Hottest:", hottest)
print("Coldest:", coldest)
```

### Task 2.2.2 — Series addition with fill_value

```python
import pandas as pd

a = pd.Series([1, 2, 3, 4, 5], index=['A', 'B', 'C', 'D', 'E'])
b = pd.Series([7, 8, 9, 0, 6], index=['A', 'B', 'C', 'D', 'F'])

c = a + b
print(c)

add = a.add(b, fill_value=0)
print(add)
```

### Task 2.2.3 — Filter values above mean

```python
import pandas as pd

s = pd.Series([45, 82, 67, 55, 91, 38, 74])

mean = s > s.mean()

print("Above Mean:\n", s[mean])
```

### Task 3.3.3 — describe(), highest std, most skewed column

```python
import seaborn as sns

df = sns.load_dataset('titanic')

desc = df.describe()

highest = desc.loc['std'].idxmax()
highestv = desc.loc['std'].max()

print("Highest STD:", highest)
print("Highest STD value:", highestv)

skew = df.skew(numeric_only=True)
most_skewed = skew.abs().max()

print("Skewed:\n", skew)
print("Most Skewed:", most_skewed)
```

### Task 3.3.4 — df_summary() utility function

```python
import pandas as pd
import seaborn as sns

def df_summary(df):
    print("Shape of DataFrame:", df.shape)
    print("Column Names:", df.columns)
    print("\nData Types:\n", df.dtypes)
    print("\nFirst 5 rows:\n", df.head(3))
    print("\nMissing Values:\n", df.isnull().sum())

titanic = sns.load_dataset("titanic")
df_summary(titanic)
```

## Section 5 — Missing Values

### Task 5.2.1 — % missing, drop >50%, fill age by pclass median

```python
import seaborn as sns

df = sns.load_dataset('titanic')

missing_percent = df.isnull().sum() / len(df) * 100
print(missing_percent)

df = df.loc[:, missing_percent < 50]
print(df.columns)

df['age'] = df.groupby('pclass')['age'].transform(
    lambda x: x.fillna(x.median())
)

print(df['age'].isnull().sum())
```

### Task 5.2.2 — dropna, fillna(0), fillna(mean), interpolate

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'A': [1, np.nan, 3, np.nan],
    'B': [5, 6, np.nan, 8]
})

print("Original:")
print(df)

print("\nDropna:")
print(df.dropna())

print("\nFill 0:")
print(df.fillna(0))

print("\nFill Mean:")
print(df.fillna(df.mean()))

print("\nInterpolate:")
print(df.interpolate())
```

### Task 5.2.3 — Fill numeric with mean, categorical with mode

```python
import seaborn as sns

df = sns.load_dataset('titanic')

df = df.fillna(df.mean(numeric_only=True))
df = df.fillna(df.mode().iloc[0])

print(df.isnull().sum())
```

## Section 6 — Feature Engineering & Cleaning

### Task 6.1.1 — FamilySize, IsAlone, AgeGroup

```python
import seaborn as sns

df = sns.load_dataset('titanic')

df['FamilySize'] = df['sibsp'] + df['parch'] + 1
df['IsAlone'] = (df['FamilySize'] == 1).astype(int)
df['AgeGroup'] = df['age'].apply(
    lambda x: 'Child' if x < 18 else 'Senior' if x > 60 else 'Adult'
)

print(df[['FamilySize', 'IsAlone', 'AgeGroup']].head())
```

## Section 7 — Comprehensive Challenges

### Challenge 7.1 — Student grade report

```python
import pandas as pd
import numpy as np

students = pd.DataFrame({
    'Name': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],
    'Math': np.random.randint(40, 100, 10),
    'Science': np.random.randint(40, 100, 10),
    'English': np.random.randint(40, 100, 10),
    'History': np.random.randint(40, 100, 10)
})

students['Total'] = students[['Math', 'Science', 'English', 'History']].sum(axis=1)
students['Average'] = students['Total'] / 4

def grade(avg):
    if avg >= 80: return 'A'
    elif avg >= 65: return 'B'
    elif avg >= 50: return 'C'
    else: return 'F'

students['Grade'] = students['Average'].apply(grade)

sorted_df = students.sort_values(by='Average', ascending=False)
print(sorted_df)
```

### Challenge 7.2 — Titanic full cleaning & analysis

```python
import seaborn as sns

df = sns.load_dataset('titanic')

df['age'] = df['age'].fillna(df['age'].median())
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
df['deck'] = df['deck'].fillna('Unknown')

df['FamilySize'] = df['sibsp'] + df['parch'] + 1
df['IsAlone'] = (df['FamilySize'] == 1).astype(int)

children = df[(df['age'] < 18) & (df['survived'] == 1) & (df['pclass'] == 3)]
print(children[['age', 'sex', 'pclass', 'survived']].head())

women_1st = df[(df['sex'] == 'female') & (df['pclass'] == 1)]
survival_percent = women_1st['survived'].mean() * 100
print("Survival %:", survival_percent)
```

### Challenge 7.3 — Sales DataFrame analysis

```python
import pandas as pd
import numpy as np

np.random.seed(1)

sales = pd.DataFrame({
    'Date': pd.date_range('2025-01-01', periods=20),
    'Product': np.random.choice(['Laptop', 'Phone', 'Tablet'], 20),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], 20),
    'Units': np.random.randint(1, 50, 20),
    'Price': np.random.randint(100, 500, 20)
})

sales['Revenue'] = sales['Units'] * sales['Price']

high_revenue = sales[sales['Revenue'] > 5000]
print(high_revenue)

top_products = sales.groupby('Product')['Revenue'].sum().sort_values(ascending=False).head(3)
print(top_products)

print(sales.isnull().sum())
```
